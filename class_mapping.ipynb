```python
import os
import json

# path ke folder train
dataset_path = "DatasetBatikFix/train"

# ambil nama folder (kelas)
classes = sorted(os.listdir(dataset_path))

# filter hanya folder
classes = [cls for cls in classes if os.path.isdir(os.path.join(dataset_path, cls))]

# buat mapping
class_indices = {cls_name: idx for idx, cls_name in enumerate(classes)}

# simpan ke JSON
with open("class_mapping.json", "w") as f:
    json.dump(class_indices, f, indent=4)

print("Class mapping berhasil dibuat!")
print(class_indices)
```
